```python
import pandas as pd

df=pd.read_excel("data/gunluk_veri.xlsx",
                usecols = "C:P")
```

```python
for i in range(len(df["Günlük Vaka"])):
    if i==0 :
        df["Günlük Vaka"][i]=df["Toplam Vaka"][i]
    else:
        df["Günlük Vaka"][i]=df["Toplam Vaka"][i]-df["Toplam Vaka"][i-1]
```
